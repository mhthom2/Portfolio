# Analysis of Chicago Cub Hitting: Daytime vs Nighttime

The Chicago Cubs are a Major League Baseball (MLB) team that's known for its iconic ballpark: Wrigley Field. One defining characteristic of Wrigley Field that didn't have lights installed until August 1988, so all of their games until then were played during the day. And even since then, the number of day games is limited because of public pressure from the neighborhoods around the ball park. In 2023, the team only played 90 night games out of 162 games, which was second fewest in the league. 

Because the Cubs play about 44% of their games during the day, an analyst might expect their players to perform well during the day. And according to my exploratory analysis of the 2023 season, using the team's batting average, on-base percentage plus slugging percentage, and just slugging percentage, the team seemed to perform worse during the day. However, using hypothesis testing, the batting averages for the daytime were greater than or equal to the averages at night.    

```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import shapiro, levene, ttest_ind
```

```python
# Set up empty data frame with column names
# BA: Batting Average, OBP: On-Base Percentage, SLG: Slugging Percentage
# OPS: On-Base Percentage + Slugging Percentage, TB: Total Bases 
cols = ['Team', 'Day/Night', 'Games', 'BA', 'OBP', 'SLG', 'OPS', 'TB']
team_data = pd.DataFrame(columns=cols)
```

```python
# Abbreviations for all teams 
teams = ['ARI', 'ATL', 'BAL', 'BOS', 'CHC', 'CHW', 
         'CIN', 'CLE', 'COL', 'DET', 'HOU', 'KCR', 
         'LAA', 'LAD', 'MIA', 'MIL', 'MIN', 'NYM', 
         'NYY', 'OAK', 'PHI', 'PIT', 'SDP', 'SEA',
         'SFG', 'STL', 'TBR', 'TEX', 'TOR', 'WSN']
```

```python
# Links to data on Stathead web site, MLB games in 2023
# Day games: https://stathead.com/tiny/6NtCP 
# Night games: https://stathead.com/tiny/QolvV
```

```python
# Process files for each team 
for team in teams:
    
    # File names for day / night data
    day_file = './assets/' + team + '_day.csv'
    night_file = './assets/' + team + '_night.csv'
    # Read in data for day / night 
    day = pd.read_csv(day_file)
    night = pd.read_csv(night_file)
    
    # Add new row to data frame with day data 
    new_row_1 = pd.Series([day.loc[0, 'Team'],
                           day.loc[0, 'Day/Night'],
                           len(day), 
                           round(np.mean(day['BA']), 3), 
                           round(np.mean(day['OBP']), 3),
                           round(np.mean(day['SLG']), 3),
                           round(np.mean(day['OPS']), 3),
                           round(np.mean(day['TB']), 2)])
    new_row_1 = new_row_1.set_axis(cols)
    team_data = pd.concat([team_data, 
                           new_row_1.to_frame().T], 
                          ignore_index=True)
    
    # Add new row to data frame with night data 
    new_row_2 = pd.Series([night.loc[0, 'Team'],
                           night.loc[0, 'Day/Night'],
                           len(night), 
                           round(np.mean(night['BA']), 3), 
                           round(np.mean(night['OBP']), 3),
                           round(np.mean(night['SLG']), 3),
                           round(np.mean(night['OPS']), 3),
                           round(np.mean(night['TB']), 2)])
    new_row_2 = new_row_2.set_axis(cols)
    team_data = pd.concat([team_data, 
                           new_row_2.to_frame().T], 
                          ignore_index=True)
```

```python
# Sorting by 'Games' to compare teams with most/least daytime games 
team_data_sorted = team_data.sort_values(by='Games').reset_index(drop=True)
# Teams with fewest day games
team_data_sorted.head()
```

```python
# Split 30 teams into 3 ten-team groups by day games 
bottom_10 = team_data_sorted.iloc[0:10]
middle_10 = team_data_sorted.iloc[10:20]
top_10 = team_data_sorted.iloc[20:30]
```

```python
# Append night game information to daytime data 
top_10 = pd.concat([top_10, team_data_sorted.iloc[30:40]], 
                   ignore_index=True)
middle_10 = pd.concat([middle_10, team_data_sorted.iloc[40:50]], 
                      ignore_index=True)
bottom_10 = pd.concat([bottom_10, team_data_sorted.iloc[50:60]], 
                             ignore_index=True)
```

```python
# Teams with most day games  
top_10 = top_10.sort_values(by='Team').reset_index(drop=True)
top_10.head(6)
```

```python
# Middle third of teams with most day games  
middle_10 = middle_10.sort_values(by='Team').reset_index(drop=True)
middle_10.head(6)
```

```python
# Teams with least  day games
bottom_10 = bottom_10.sort_values(by='Team').reset_index(drop=True)
bottom_10 = bottom_10.sort_values('Games').reset_index(drop=True)
bottom_10.head(6)
```

```python
# Calculate percentage for day games for Chicago Cubs 
cubs_percent_day = round(top_10.loc[0, 'Games'] / (top_10.loc[0, 'Games'] + top_10.loc[1, 'Games']), 4)

# Same for LA Dodgers, team with fewest day games 
dodgers_percent_day = round(bottom_10.loc[0, 'Games'] / \
                      (bottom_10.loc[0, 'Games'] + bottom_10.loc[len(bottom_10)-1, 'Games']), 4)

# Print results 
print('Percentage of Cub day games: ', cubs_percent_day)
print('Percentage of Dodger day games: ', dodgers_percent_day)
```

```python
# Get data for Chicago Cubs 
cubbies = top_10[0:2].drop(['Team', 'Games', 'TB', 'OPS'], axis=1)

# Wrangle data into plotable form 
cubbies = cubbies.iloc[:, 1:5].T
stats = pd.Series(['BA', 'OBP', 'SLG', 'BA', 'OBP', 'SLG'], name='Stats')
day_night = pd.Series(['Day', 'Day', 'Day', 'Night', 'Night', 'Night'], 
                      name='Day/Night')
cubbies = pd.DataFrame(pd.concat([cubbies[0], cubbies[1]], ignore_index = True))
cubbies = pd.concat([stats, day_night, cubbies], axis=1)
cubbies = cubbies.rename({0: 'Value'}, axis=1)
cubbies
```

```python
# Plot data using seaborn module 

# Set figure size 
sns.set(rc={"figure.figsize":(6, 4)})
# Plot data 
chart = sns.barplot(x = 'Stats', y = 'Value', 
                    hue = 'Day/Night', data = cubbies,
                    palette='Set1')
# Set plot title 
chart.set_title('Chicago Cubs Hitting Statistics in 2023 Daytime vs Nighttime')
# Add x-tick labels 
chart.set_xticks(ticks=[0, 1, 2], 
                 labels=['Batting\nAverage', 'On Base\nPercentage', 'Slugging\nPercentage'])
# Label x data 
chart.set_xlabel('Hitting Statistics')
# Label y data 
chart.set_ylabel('Statistical Value')
# Remove legend title 
chart.legend(title='')

# To save plot
fig = chart.get_figure()
fig.savefig('Cubs_Hitting_Day_v_Night.png', bbox_inches='tight');
```

From the bar charts, it seems that the team batting statistics were better during the day than at night.

To confirm this, we did hypothesis testing on one of the averages, OPS. 

### Hypothesis Testing on OPS 
1. Define Hypothesis
   
   Let $\mu_D = $ mean of OPS during day, and $\mu_N = $ mean of OPS at night
   
   $H_0$: $\mu_N \le \mu_D$  The mean of the OPS during the day is the same or greater than the mean of the OPS at night. 
   
   $H_1$: $\mu_D \gt \mu_N$  The mean of the OPS at night is greater than the OPS during the day. 

```python
# Get Cubs day data 
CHC_day_file = './assets/CHC_day.csv'
CHC_night_file = './assets/CHC_night.csv'

# Read in data for day / night 
CHC_day = pd.read_csv(day_file)
CHCnight = pd.read_csv(night_file)
```

2. Assumption Check: Normality of Data

   $H_0$: The data is normally distributed.

   $H_1$: The data isn't normally distributed.

Assuming $\alpha = 0.05$. 

```python
# Using Shapiro-Wilks W test to check normality
day_shapiro, day_p_value = shapiro(CHC_day['OPS'])
night_shapiro, night_p_value = shapiro(CHC_night['OPS'])

print('Day p-value: ', day_p_value)
print('Night p-value: ', night_p_value)
```

Therefore, data meets the assumption that the data is normally distributed. 

For most t-testing, we would also check the variance of the data. However, the number of daytime 
samples is not the same as the number of nighttime samples. So we can't compare the variances, 
and won't use the student t-test that requires the variance of the samples be the same. Instead, 
we'll use Welch's t-test, which doesn't have that requirement. 

```python
# Test hypothesis using paired t-test
ttest, t_p_value = ttest_ind(CHC_night['OPS'], CHC_day['OPS'], 
                             alternative='greater', equal_var=False)
print('t-test p-value: ', t_p_value)
```

Therefore, the null hypothesis can't be rejected, and the OPS during the day are statistically greater than 
or equal to the averages at night. So even though the bar charts seem to indicate that the OPS was better at than during the day, the difference isn't statistically significant.
